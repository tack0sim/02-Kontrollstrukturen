>## **02.3 <span style="color:#fcba03">Quest 1</span>** Loop 1 bis 10 addieren

Erstellen Sie ein kleines Programm, das mit Hilfe einer While Schleife\
die Zahlen 1 bis 10 addiert und jede Iteration auf dem Bildschirm ausgibt.

<div class="alert alert-block alert-success">
<b>02.3 Lösung Quest 1 Loop 1 bis 10</b>
</div>

```python
let i = 0
let add = 0

while (i < 10) {
    add += i
    i++
    console.log(add)
}
```
